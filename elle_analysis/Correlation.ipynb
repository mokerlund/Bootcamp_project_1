```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
# load in clean data for 2018 and remove Pennsylvania 

file_path = "cleaned_data/final_clean_data_2018.csv"
df=pd.read_csv(file_path)
# col_drop = ["State","Time Period Date Start","Time Period Date End","State FIPS Code"]
# df=df.drop(col_drop,axis=1)
print(df.shape)

df2=df[df["County"]!="Pennsylvania"]
print(df2.shape)
df2.head()
```

```python
corr = df2.corr(method="pearson")
corr = pd.DataFrame(corr)
print(corr.columns)
corr
```

```python
# correlation b/w % of any drug OD death vs. ED visits, and intervensions 

death_rate_corr = corr.loc["%_of_any_od_death_over_all_deaths",\
                      ['Annual Any Overdose ED Rate by Visit',\
                       'Annual Any Overdose ED Rate by Population',\
                       'hotlinecall_2018',\
                       'Number Boxes']]

death_rate_corr=pd.DataFrame(death_rate_corr)
death_rate_corr=death_rate_corr.rename(columns={"%_of_any_od_death_over_all_deaths":"Correlation between overall OD death and intervensions"})
death_rate_corr
```

```python
# correlation b/w % of opioid OD death vs. ED visits, and intervensions 

death_rate_corr_2 = corr.loc["%_of_op_od_death_over_all_deaths",\
                      ['Annual Opioid Overdose ED Rate by Visit',\
                       'Annual Opioid Overdose ED Rate by Population',\
                       'hotlinecall_2018',\
                       'Number Boxes']]

death_rate_corr_2=pd.DataFrame(death_rate_corr_2)

death_rate_corr_2=death_rate_corr_2.rename(columns={"%_of_op_od_death_over_all_deaths":"Correlation between opioid OD death and intervensions"})
death_rate_corr_2
```
